```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt

def seirs_model(y, t, alpha, beta, gamma, sigma):
    S, E, I, R = y
    dSdt = -beta*S*I + gamma*R
    dEdt = beta*S*I - sigma*E
    dIdt = sigma*E - alpha*I
    dRdt = alpha*I - gamma*R
    return [dSdt, dEdt, dIdt, dRdt]

# Define parameters
beta = 0.32
sigma = 0.87
alpha = 0.15
gamma = 0.69

# Set initial conditions
S0 = 9785
E0 = 115
I0 = 100
R0 = 0

# Set time span
tspan = np.arange(0, 100, 0.01)

# Solve the differential equations
y0 = [S0, E0, I0, R0]
sol = odeint(seirs_model, y0, tspan, args=(alpha, beta, gamma, sigma))

# Extract solutions
S, E, I, R = sol.T

# Plot the results
plt.plot(tspan, S, label='Susceptible')
plt.plot(tspan, E, label='Exposed')
plt.plot(tspan, I, label='Infectious')
plt.plot(tspan, R, label='Recovered')
plt.xlabel('Time')
plt.ylabel('Population')
plt.legend()
plt.show()
```
